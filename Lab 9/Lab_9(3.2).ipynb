```python
#Элемент матрицы называется локальным минимумом, если он строго меньше всех
#имеющихся у него соседей. Подсчитать количество локальных минимумов заданной
#матрицы размером 10 на 10. Найти сумму модулей элементов, расположенных выше
#главной диагонали.

#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import numpy as np 

nrows = 10
ncols = 10

m = np.random.randint(25, size=(nrows, ncols))

print(m)
```

```python
count = 0
for i in range(nrows):
    for j in range(ncols):
        if i == 0 and j == 0:
            if m[i][j] <= m[i][j+1] and  m[i][j] <= m[i+1][j]:
                count += 1
                print(f"m[{i}][{j}] = {m[i][j]}")
        elif i == 0 and j == nrows-1:
            if m[i][j] <= m[i][j-1] and  m[i][j] <= m[i+1][j]:
                count += 1
                print(f"m[{i}][{j}] = {m[i][j]}")
        elif i == ncols-1 and j == 0:
            if m[i][j] <= m[i][j+1] and  m[i][j] <= m[i-1][j]:
                count += 1
                print(f"m[{i}][{j}] = {m[i][j]}")
        elif i == ncols-1 and j == nrows-1:
            if m[i][j] <= m[i][j-1] and  m[i][j] <= m[i-1][j]:
                count += 1
                print(f"m[{i}][{j}] = {m[i][j]}")
        elif i == 0:
            if m[i][j] <= m[i+1][j] and  m[i][j] <= m[i][j-1] and  m[i][j] <= m[i][j+1]:
                count += 1
                print(f"m[{i}][{j}] = {m[i][j]}")
        elif i == ncols-1:
            if m[i][j] <= m[i][j+1] and  m[i][j] <= m[i-1][j] and  m[i][j] <= m[i][j-1]:
                count += 1   
                print(f"m[{i}][{j}] = {m[i][j]}") 
        elif j == 0:
            if m[i][j] <= m[i+1][j] and  m[i][j] <= m[i-1][j] and  m[i][j] <= m[i][j+1]:
                count += 1
                print(f"m[{i}][{j}] = {m[i][j]}")
        elif j == nrows-1:
            if m[i][j] <= m[i-1][j] and  m[i][j] <= m[i][j-1] and  m[i][j] <= m[i+1][j]:
                count += 1
                print(f"m[{i}][{j}] = {m[i][j]}")
        else:
            if m[i][j] <= m[i][j+1] and  m[i][j] <= m[i+1][j] and  m[i][j] <= m[i][j-1] and  m[i][j] <= m[i-1][j]:
                count += 1
                print(f"m[{i}][{j}] = {m[i][j]}")
print(f"Кол-во локальных минимумов:{count}")
```
